```kotlin
%%kotlin
object GuildWarehouseLevel {
    fun capacityForLevel(level: Int): Int = 25 + ((level - 1) * 25)
    fun goldCostForLevel(level: Int): Long = 1_500_000L * (1 shl (level - 1))
    fun effectForLevel(level: Int): String = "Capacity: ${capacityForLevel(level)} slots"

    fun generateMarkdownTable(maxLevel: Int): String {
        val sb = StringBuilder()
        sb.append("| Level | Gold Cost | Effect |\n")
        sb.append("|------:|-----------:|:-------|")
        for (lvl in 1..maxLevel) {
            sb.append("\n| $lvl | ${goldCostForLevel(lvl)} | ${effectForLevel(lvl)} |")
        }
        return sb.toString()
    }
}

// Prompt user for max guild level
val input = window.prompt("Enter max guild level:", "10")
val maxLevel = input?.toIntOrNull() ?: 10

GuildWarehouseLevel.generateMarkdownTable(maxLevel)
```
